```python
import matplotlib
matplotlib.rcParams.update({'font.size': 18})
from matplotlib import pyplot as plt, cm
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from matplotlib.font_manager import FontProperties
font = FontProperties()
font.set_size('x-large')
```

# Analisis

# Fuerza Bruta
### - FB solo
### - FB vs 2^n
### - FB correlacion

```python
random_fb = pd.read_csv('resultados/random_suite_res_fb_2.txt', delimiter=' ')
random_fb['ms'] = random_fb['t'] / 1000000
random_fb['esperado'] = (2 ** random_fb['n']) * 19 / 1000000
```

```python
plt.figure(figsize=(10, 7))
fig = sns.lineplot(data=random_fb.groupby(['n','r']).mean().reset_index(), x='n', y='ms')
fig.legend(labels=['FB'])
fig.set(title="Tiempo de ejecucion", xlabel='n', ylabel='tiempo (ms)')
plt.savefig('img/fb_solo')
```

```python
plt.figure(figsize=(10, 7))
fig = sns.lineplot(data=random_fb.groupby(['n','r']).mean().reset_index(), x='n', y='esperado', color='orange',ci=None)
fig = sns.scatterplot(data=random_fb.groupby(['n','r']).mean().reset_index(), x='n', y='ms', s=matplotlib.rcParams['lines.markersize'] ** 2)
fig.legend(labels=['O(2^n)','FB'])
fig.set(title="Tiempo de ejecucion y esperado", xlabel='n', ylabel='tiempo (ms)')
plt.savefig('img/fb_vs_esp')
```

```python
plt.figure(figsize=(10, 7))
fig_fb = sns.lineplot(data=random_fb.groupby(['n','r']).mean().reset_index(), x='esperado', y='esperado', color='orange',ci=None)
fig_fb = sns.scatterplot(data=random_fb.groupby(['n','r']).mean().reset_index(), x='esperado', y='ms')
fig_fb.set(title="Correlacion entre Tiempo de ejecucion y esperado", xlabel='esperado', ylabel='tiempo (ms)')
plt.savefig('img/fb_corr')
```

# BT
### - BTO vs BTF
### - BTO vs BTF vs FB
### - BTO vs BTF vs FB (Peor caso Fact)
### - BTO vs BTO FLP

```python
random_btf = pd.read_csv('resultados/random_suite_res_btf_2.txt', delimiter=' ')
random_btf['ms'] = random_btf['t'] / 1000000
random_bto = pd.read_csv('resultados/random_suite_res_bto_2.txt', delimiter=' ')
random_bto['ms'] = random_bto['t'] / 1000000
```

```python
plt.figure(figsize=(10, 7))
fig = sns.lineplot(data=random_btf.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig = sns.lineplot(data=random_bto.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig.legend(labels=['BTF','BTO'])
fig.set(title="Tiempo de ejecucion de BTO y BTF", xlabel='n', ylabel='tiempo (ms)')
plt.savefig('img/bto_vs_btf')
```

```python
plt.figure(figsize=(10, 7))
fig = sns.lineplot(data=random_fb.groupby(['n','r']).mean().reset_index(), x='n', y='ms')
fig = sns.lineplot(data=random_btf.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig = sns.lineplot(data=random_bto.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig.legend(labels=['FB','BTF','BTO'])
plt.ylim(-0.001, 2)
fig.set(title="Tiempo de ejecucion de BTO, BTF, y FB", xlabel='n', ylabel='tiempo (ms)')
plt.savefig('img/bto_vs_btf_vs_fb')
```

## Worse case

```python
worse_bto = pd.read_csv('resultados/wc_bto.txt', delimiter=' ')
worse_bto['ms'] = worse_bto['t'] / 1000000
worse_btf = pd.read_csv('resultados/wc_btf.txt', delimiter=' ')
worse_btf['ms'] = worse_btf['t'] / 1000000
worse_bto_flp = pd.read_csv('resultados/wc_bto_flp.txt', delimiter=' ')
worse_bto_flp['ms'] = worse_bto_flp['t'] / 1000000
```

```python
plt.figure(figsize=(10, 7))
fig = sns.lineplot(data=worse_bto_flp.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig = sns.lineplot(data=worse_bto.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig.legend(labels=['BTO INV','BTO'])
plt.ylim(-0.001, 0.008)
fig.set(title="Peor Caso Factibilidad Sobre: BTO y BTO Inv", xlabel='n', ylabel='tiempo (ms)')
plt.savefig('img/bto_vs_bto_flp')
```

```python
plt.figure(figsize=(10, 7))
fig = sns.lineplot(data=worse_fb.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig = sns.lineplot(data=worse_btf.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig = sns.lineplot(data=worse_bto.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
# fig = sns.lineplot(data=worse_bto_flp.groupby(['n','r']).mean().reset_index(), x='n', y='ms', ci=None)
fig.legend(labels=['FB','BTF','BTO'])
plt.ylim(-0.001, 1)
fig.set(title="Peor Caso Factibilidad Sobre: FB, BTF, BTO y BTO Inv", xlabel='n', ylabel='tiempo (ms)')
plt.savefig('img/fb_btf_bto_bto_flp_wc')
```

```python
plt.figure(figsize=(10, 7))
fig = sns.lineplot(data=worse_fb.groupby(['n','r']).mean().reset_index()[:10], x='n', y='ms', ci=None)
fig = sns.lineplot(data=worse_btf.groupby(['n','r']).mean().reset_index()[:10], x='n', y='ms', ci=None)
fig.legend(labels=['FB','BTF'])
plt.ylim(-0.001, 0.1)
fig.set(title="Peor Caso Factibilidad Sobre: FB, BTF", xlabel='n', ylabel='tiempo (ms)')
plt.savefig('img/btf_vs_fb_wc')
```

```python
stress_pd = pd.read_csv('resultados/results_pd_exahustivo.txt', delimiter=' ')
stress_pd['ms'] = stress_pd['t'] / 1000000
stress_pd['esperado'] = (stress_pd['n'] * stress_pd['r'] * 1.28 / 100000)-60
```

```python
plt.figure(figsize=(10, 7))
fig = sns.scatterplot(data=stress_pd.groupby(['n','r']).mean().reset_index().loc[stress_pd.r==10000], x='n', y='ms')
fig = sns.lineplot(data=stress_pd.groupby(['n','r']).mean().reset_index().loc[stress_pd.r==10000], x='n', y='esperado', ci=None)
fig.legend(labels=['PD','PD Esperado'])
fig.set(title="Correlacion entre Tiempo de ejecucion y esperado", xlabel='n', ylabel='tiempo (ms)')
plt.savefig('img/pd_corr')
```
